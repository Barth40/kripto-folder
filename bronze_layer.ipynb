```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
import pandas as pd
import requests
```

```python
import requests 

url = "https://api.coingecko.com/api/v3/coins/markets"

params = {
    "vs_currency": "usd",
    "order": "market_cap_desc",
    "per_page": 250,
    "page": 1,
}

response = requests.get(url=url, params=params)

if response.status_code == 200:
    print('Successfully connected')
else:
    print(f'Error due to: {response.status_code}')
```

```python
print(response.status_code)
```

```python
if response.status_code == 200:
    data = response.json()
    print(data[:2])  # Print just the first 2 entries to keep it readable
else:
    print(f'Error due to: {response.status_code}')
```

```python
if response.status_code == 200:
    data = response.json()
    if data:
        keys = data[0].keys()  # Get keys from the first dictionary in the list
        print(list(keys))
    else:
        print("No data returned.")
else:
    print(f'Error due to: {response.status_code}')
```

```python
pdf = pd.DataFrame(data)          # Create pandas DataFrame
df = spark.createDataFrame(pdf)   # Convert pandas DataFrame to Spark DataFrame
```
